```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("../media/data_2.csv")
df
```

```python
plt.xlabel('Area(sqr ft)')
plt.ylabel('Price(US Dollars)')
plt.scatter(df.area, df.precio, color='red',marker='+')
```

```python
#Entrenar el Modelo con regresin lineal
reg = LinearRegression()
reg.fit(df[['area']], df.precio)
```

```python
#Probar el Modelo
df
reg.predict([[3100]])
```
